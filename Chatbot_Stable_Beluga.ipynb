# Imports

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from IPython.display import clear_output
import time
```

# Varialble

```python
write_question = '\nWrite your question.'
quit_method_message = 'For quit from the chat write "quit" in the box:\n'
input_message = f" {write_question} {quit_method_message}"
quit_message = '\n*** Thank you for conversation. See you next time! ***'
next_quote_message = '\n*** Write your next prompt. I will try to answer for it ***\n'
quote_separator_message = f'-'*70
patient_message = 'The answer is being generated, please be patient...'
```

# Functions

```python
def chat_with_bot(message, tokenizer, model):
    print(f'\n{patient_message}')
    user_message = f'\n### User quote ###\n {message}'
    prompt = f"{user_message} \n\n### CHATBOT ###\n"
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    output = model.generate(**inputs, do_sample=True, top_p=0.95, top_k=0,
                            max_new_tokens=256)
    clear_output(wait=True)

    print(tokenizer.decode(output[0], skip_special_tokens=True))
    print(f'{next_quote_message}{quote_separator_message}')


def initiate_chat(input_message, quit_message, tokenizer, model):
    while True:
        message = input(input_message)

        if message.lower().strip() == 'quit':
            print(quit_message)
            break

        chat_with_bot(message, tokenizer, model)


def start_conversation():
    input_message = f"{write_question} {quit_method_message}"

    initiate_chat(input_message, quit_message, tokenizer, model)
```

# Code

```python
tokenizer = AutoTokenizer.from_pretrained("stabilityai/StableBeluga-7B", use_fast=False)
model = AutoModelForCausalLM.from_pretrained("stabilityai/StableBeluga-7B", torch_dtype=torch.float16, low_cpu_mem_usage=True, device_map="auto")
```

```python
start_conversation()
```
